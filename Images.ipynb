```python
import pandas as pd
import requests
from PIL import Image
```

```python
file=pd.read_csv("list1.csv")
```

```python
file.head()
```

```python
link=file['Image']
```

```python
for i in link:
    try:
        file_name = i.split('/')[-1]
        print("Downloading file:%s"%file_name)
        r = requests.get(i, stream=True)
        # this should be file_name variable instead of "file_name" string
        with open(file_name, 'wb') as f:
            for chunk in r:
                f.write(chunk)
    except:
        print("Error")
```
